```python
#1 - Realizar a média da metragem para cara um dos bairros (Paulo)
media_metragem_bairro = dados.loc[:,["Bairro","Metragem"]].groupby(by = "Bairro").mean().reset_index().sort_values("Metragem")
media_metragem_bairro
```

```python
#2 - 2 formas de selecionar os dados por bairro (consultar os métodos na documentação do pandas) (Thiago)
seleciona_bairro1 = dados[dados["Bairro"] == "Alto da Mooca"]
seleciona_bairro2 = dados.loc[dados["Bairro"]=="Alto da Mooca"]
sel_bairro = ["Vila Mariana","Alto da Mooca"]
seleciona_bairro3 = dados[dados["Bairro"].isin(sel_bairro)]
```

```python
#3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)
#Bar plot the mediana de quartos por Bairro
quarto_bairro = dados.loc[:,["Bairro","Quartos"]].groupby("Bairro").median().sort_values("Quartos",ascending = False)
quarto_bairro.head(10).plot(kind = "barh")
```

O bairro com a maior mediana de quatos é o Jardim Gonzaga com 10 quartos na mediana. Depois seguem Vila Picinin com 9, Vila Sabrina com 8 e Jardim Japao com 7. Todos dos outros bairros tem medianada de 6 ou menos quartos.

```python
#Scatter plot Banheiros Vs Metragem
import matplotlib.pyplot as plt
x = dados['Banheiros']
y = dados["Metragem"]
ax = plt.scatter(x,y)
plt.xlabel("Numero de Banheiros")
plt.ylabel("Metragem")
```

Nao existe correlacao aparente entre o numero de banheiros e a metragem de um imovel.

```python
#4 - Pegar outros estatísticas dos dados, média, mediana, mim, max (Vivi)
dados.describe()
```

```python
dados.describe().transpose()['std']*100/dados.describe().transpose()['mean']
```

A media é significativamente maior que a mediana da Metragem e dos numero de Quartos indicando que os esses dados talvez nao tenham uma distribuicao simetrica. Isso se da possivelmente pela presenca de imoveis muito grandes e com muitos quartos que elevam o valor da media.
O numero de banheiros de uma propriedade varia mais do que o numero de quartos mas a maior variacao esta no tamanho do imovel.
O maior imovel tem 750 000 metros quadrados enquanto o menos tem somente 10. Metade dos imoveis tem tamanho igual ou menos do que 290 metros quadrados. 
Tres quartos dos imoveis tem ao menos 4 quartos. Todos os imoveis tem ao menos 1 banheiro.
