- Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)
- Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
#Criar uma escala de valores em milhões
dados_vendas["Valor_anuncio_Milhoes"] = dados_vendas["Valor_anuncio_float"].divide(1000000)
dados_vendas.describe()
```

```python
#Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)
sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_Milhoes", bins = 200,kde = True, color = "green")
ax.set_title("Histograma Valor Imóvel")
ax.set_xlabel("Valor do Anuncio em Milhoes")
ax.set_ylabel("Numero de Imoveis")
plt.xlim((-0.1, 10))
plt.xticks(list(range(0,11)))
plt.show()
```

```python
#Preço do metro quadrado por bairro e plotar em um gráfico ideal
dados_vendas["valor_metro"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"] 
valor_metragem_bairro = dados_vendas[['Bairro','valor_metro']].groupby('Bairro').mean().reset_index().sort_values(by='valor_metro', ascending = False)
ax = sns.barplot(y=valor_metragem_bairro.head(10)['Bairro'],x=valor_metragem_bairro.head(10)['valor_metro'],orient = 'h',color='blue')
ax.set_title("Valor medio o metro - 10 mais caros")
ax.set_xlabel("Valor por metro")
ax.set_ylabel("Bairro")
plt.show()
```

Explorar as bibliotecas de visualizações e colocar as suas conclusão:

As bibliotecas do pandas e do seaborn, parecem derivar do Matplotlib. Existem muitas possibilidades com o Matplotlib em si mas se alguns outros efeitos sao desejados e de modo mais simples melhor aplicar as outras bibliotecas.
A biblioteca ggplot vem da linguagem R e parece ser simples de implementar.
Existem outras bibliotecas de uso mais especifico para interacao como o Bokeh e oPlotly

```python
#Pesquisar um visualização para analisar os quartis, mediana e outliers
plt.figure(figsize=(8, 12))
ax = sns.boxplot(y = 'Banheiros',data = dados_vendas)
ax.set_title("Distribuicao do numero de banheiros")
ax.set_xlabel("Banheiros")
ax.set_ylabel("Numero de banheiros")
plt.show()
```
